```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"] = 'True'
os.environ["TF_CPP_MIN_LOG_LEVEL"] = '2'
```

```python
import random
import tensorflow as tf
seedval = 20
tf.random.set_seed(seedval)
np.random.seed(seedval)
random.seed(seedval)
os.environ["PYTHONHASHSEED"] = str(seedval)
```

```python
df = pd.read_csv("data/points_nand.csv")
data_train = df[["x1", "x2"]]
class_train = df["class"]
print(data_train.shape)
print(class_train.shape)
```

### ニュートラルネットワークの定義
##### Sequentialクラスを使用してニュートラルネットワークの構造を定義  
##### .add：modelに各層のニューロンを追加  
##### Dense：ニューロンを配置する関数
- Dense(5, input_shape=(2,), activation="relu")
  - 第1引数: 中間層となるニューロンの数
  - input_shape: 入力の数
  - activation: 使用する活性化関数
      - sigmoid : シグモイド関数
      - relu : ReLU関数

```python
model = Sequential()
model.add(Dense(5, input_shape=(2,), activation="relu"))
# ↓一つ前の全ニューロンと全結合しているのでinput_shapeを指定しなくても良い。
model.add(Dense(1, activation="sigmoid"))　
```

### トレーニング
#### compileで学習用の設定を行う
- optimizer: 重み係数（$w$）と闘値（$\theta$）を最適化する方法
  - sgd (確率的勾配降下法)
  - adam (適応的モーメント推定)
- loss: 教師信号と出力のずれを測る誤差関数　例 binary_crossentropy
- metrics:学習過程で表示される精度 accuracy
#### 誤差が小さくなるように重み$w$と闘値$\theta$を最適化
- model.fit(data_train, class_train, epochs=1000, batch_size=1)  
  - data_train: 入力値　
  - class_train: 教師信号
  - epochs: 反復更新回数
  - batch_size: 入力データの個数

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
fit_log = model.fit(data_train, class_train, epochs=3000, batch_size=1)
```

### 精度のトレーニング

```python
plt.plot(fit_log.history["accuracy"], label="train")
plt.title("model accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epochs")
plt.legend()
plt.show()
```

### 予測
学習済みmodelを使用して予測値を算出する

```python
pred = model.predict(data_train)
print(pred)
```

## (発展)分離超平面の表示

### 分離超平面の設定(教師信号)

```python
t1 = data_train[class_train==1]
t0 = data_train[class_train==0]

t1_x = t1.iloc[:, 0]
t1_y = t1.iloc[:, 1]

t0_x = t0.iloc[:, 0]
t0_y = t0.iloc[:, 1]
```

### 分離超平面の設定(grid point)

```python
g = pd.read_csv("data/grid01_21x21.csv", header=None)
pred_g = model.predict(g)[:, 0]

g1 = g[pred_g>=0.5]
g0 = g[pred_g<0.5]

g1_x = g1.iloc[:, 0]
g1_y = g1.iloc[:, 1]

g0_x = g0.iloc[:, 0]
g0_y = g0.iloc[:, 1]
```

### 分離超平面

```python
plt.scatter(t1_x, t1_y, marker="s", facecolor="none", edgecolor="black", s=180)
plt.scatter(t0_x, t0_y, marker="o", facecolor="none", edgecolor="black", s=180)

plt.scatter(g1_x, g1_y, marker='$1$')
plt.scatter(g0_x, g0_y, marker='$0$')

plt.gca().set_aspect('equal', adjustable='box')
plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)

plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
