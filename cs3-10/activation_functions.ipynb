### 活性化関数のプログラム

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-6.0, 6.0, num = 100)
```

```python
def step(x):
    return (x>0).astype("int")
```

### ステップ関数

```python
plt.plot(x, step(x), label="step")
plt.xlabel("x")
plt.ylabel("Y")
plt.legend()
plt.show()
```

### シグモイド関数

```python
def sigmoid(x):
    return 1.0/(1.0 + np.exp(-x))
```

```python
plt.plot(x, sigmoid(x), label="sigmoid")
plt.xlabel("x")
plt.ylabel("Y")
plt.legend()
plt.show()
```

### ReLU関数
np.maximum(a, b)  
aとbを比較して、大きい方を採用して、新しいndarrayを返す関数

```python
def relu(x):
    return np.maximum(0, x)
```

```python
plt.plot(x, relu(x), label="relu")
plt.xlabel("x")
plt.ylabel("Y")
plt.legend()
plt.show()
```

### 恒等関数

```python
def ident(x):
    return x
```

```python
plt.plot(x, ident(x), label="ident")
plt.xlabel("x")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
plt.plot(x, ident(x), label="ident")
plt.plot(x, step(x), label="step")
plt.plot(x, sigmoid(x), label="sigmoid")
plt.plot(x, relu(x), label="relu")
plt.xlim(-6.2, 6.2)
plt.ylim(-1.2, 1.2)
plt.xlabel("x")
plt.ylabel("Y")
plt.grid()
plt.legend()
plt.show()
```
