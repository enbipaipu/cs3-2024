#### Read csv

```python
df = pd.read_csv("../cs3-02/wholesale_customers.csv", sep=",", skiprows=13, header=0)
print(df.shape)
print(df.info())
print(df.head())
```

## データの加工

```python
ser1 = 2 * df["Fresh"] + df["Frozen"]-2000
print(ser1)
```

#### データフレームへの列の追加

```python
df["new_col"] = ser1
# df["new_col"] = 2 * df["Fresh"] + df["Frozen"] - 2000
display(df.head())
```

#### 列・行の削除

```python
df_drop_c = df.drop(columns = "Detergents_Paper")
display(df_drop_c.head())
```
