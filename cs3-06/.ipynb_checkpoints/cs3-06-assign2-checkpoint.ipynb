```python
# To avoid the warning about the memory leak of KMeans (Win MKL version) 
import os
os.environ['OMP_NUM_THREADS'] = "1"

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
# To show all rows/columns
pd.options.display.max_columns = 999
pd.options.display.max_rows = 999
 
```

### ２．データをdfに読み込む

```python
df = pd.read_csv("winequality-red.csv", sep=";")
print(df.shape)
print(df.info())
display(df.head())
```

### 3.統計量の表示

```python
display(df.describe())
```

### ４．X_scaledに代入

```python
X_scaled=scale(df)
```

### 5.　X_scaledの平均と、母標準偏差

```python
print(X_scaled.mean(axis=0))
print(X_scaled.std(ddof=0, axis=0))
```

### 6.　X_scaledにdfと同じ列ラベルを付与したdf_scaledを作成

```python
df_scaled = pd.DataFrame(X_scaled, columns=df.columns)
display(df_scaled.head())
```

### 7.df_scaledにKMeans法によるElbow法を最大クラスタ数10で実施。

```python
distortions = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, n_init=10)
    km.fit(df_scaled)
    distortions.append(km.inertia_)

plt.plot(range(1, 11),distortions, marker="o")
plt.xlabel("number of cluster")
plt.ylabel("inertia")
plt.show()
```

### 8.df_scaledにクラスタリングを実行

```python
n_cls = 4

km = KMeans(n_clusters=n_cls , n_init=10, random_state=5)
cls = km.fit_predict(df_scaled)
print(cls)
```

### 9. dfにcluster_noを追加

```python
df["cluster_no"] = cls
display(df.head())
```

### 10.cluster_no列の出現数を表示

```python
print(df["cluster_no"].value_counts())
```

### 11の回答
下記の箱ひげ図より chlorides, sulphatesの二つの列がクラスタ番号0が他のクラスタと区別されるのに大k三区寄与している。

### 11. cluster_no列以外のクラスタごとの分布を箱ひげ図でか確認。

```python
cols = list(df.columns)[:-1]

for c in cols:
    dat = []
    for i in range(n_cls):
        df_cls = df[df["cluster_no"]==i]
        dat.append(df_cls[c])
    plt.boxplot(dat, labels=range(n_cls))
    plt.xlabel("cluster No.")
    plt.ylabel(c)
    plt.show()
```
