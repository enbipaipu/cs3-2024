```python
import pandas as pd 
```

```python
pd.options.display.max_columns=999
pd.options.display.max_rows = 999
```

```python
df_country = pd.read_csv("country_names.csv", sep=",")
print(df_country.shape)
print(df_country.info())
display(df_country.head())
display(df_country.tail())
```

```python
df_data = pd.read_csv("country_data2019.csv", sep=",")
print(df_data.shape)
print(df_data.info())
display(df_data.head())
display(df_data.tail())
```

### データフレームのマージ

```python
df = pd.merge(df_data, df_country, how="left", left_on="Country Code",right_on="Code")
display(df.head())
```

### 列の順序変更

```python
cols = list(df.columns)[:-2]
cols.insert(1, "Country Name")

print(cols)
```

```python
df = df[cols]
display(df.head())
```

```python
df.to_csv("country_merged.csv", index=False, encoding="utf-8-sig")
```
