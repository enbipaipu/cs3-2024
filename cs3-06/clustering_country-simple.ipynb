```python
import os 
os.environ["OMP_NUM_THREADS"] = "1"

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
```

simple version

```python
pd.options.display.max_columns = 999
pd.options.display.max_rows = 999
```

```python
df = pd.read_csv("country_merged.csv", sep=",")
print(df.shape)
print(df.info())
display(df.head())
display(df.tail())
```

```python
display(df.describe())
```

### クラスタリング用データの抽出

```python
dfx = df[["gdpp", "imports"]]
print(dfx.shape)
display(dfx.head())
```

### データの標準化

```python
x_scaled= scale(dfx)
```

```python
print(dfx.mean())
print(dfx.std(ddof=0))
print(x_scaled.mean(axis=0))
print(x_scaled.std(ddof=0,axis=0))
```

```python
print(type(x_scaled))
print(x_scaled.shape)
```

```python
dfx_scaled=pd.DataFrame(x_scaled, columns=dfx.columns)
print(type(dfx_scaled))
display(dfx_scaled.head())
```

### 散布図の描画

```python
plt.scatter(dfx_scaled["gdpp"], dfx_scaled["imports"], marker="o")
plt.xlabel("gdpp(scaled)")
plt.ylabel("imports(scaled)")
plt.show()
```

### エルボメソッド

```python
distortions = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, n_init=10)
    km.fit(dfx_scaled)
    distortions.append(km.inertia_)
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

### クラスタリングの実行

```python
n_cls=4
km=KMeans(n_clusters=n_cls, n_init=10, random_state=7)
cls=km.fit_predict(dfx_scaled)
print(cls)
```

### cluster_no列の追加

```python
df["cluster_no"] = cls
display(df.head())
```

### 各クラスタの種類数の確認

```python
print(df["cluster_no"].value_counts())
```

### 各クラスタの調査

```python
for i in range(n_cls):
    print("Cluster No.", i)
    display(df[df["cluster_no"]==i])
```

### 相関図の描写

```python
plt.scatter(df['gdpp'], df['imports'], 
            marker='o', c=df['cluster_no'])
plt.colorbar()
plt.xlabel('gdpp')
plt.ylabel('imports')
plt.show()
```

```python
print(km.cluster_centers_)
```

```python
colors = ['blue', 'green', 'orange', 'magenta']
for i in range(n_cls):
    df_cls = dfx_scaled[ df['cluster_no']==i ]
    plt.scatter(df_cls['gdpp'], df_cls['imports'],
                color=colors[i], marker='${}$'.format(i),
                label='cluster{}'.format(i))

plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
            s=250, marker='*', color='red', label='centroids')

plt.xlabel('gdpp (scaled)')
plt.ylabel('imports (scaled)')
plt.legend()
plt.show()
```
