```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import numpy as np
```

```python
df = pd.read_csv("boston_small.csv", sep=",", skiprows=8)
print(df.shape)
print(df.info())
display(df.head())
```

```python
x = df[["RM"]]
y = df["MEDV"]

print("X:", x.shape)
display(x.head())
print("y:", y.shape)
print(y.head())

x_c = sm.add_constant(x)
display(x.head())
display(x_c.head())

model = sm.OLS(y, x_c)
results = model.fit()
print(results.summary())
```

### 3.

```python
print("R2:", results.rsquared)
print("Adj R2:", results.rsquared_adj)
```

```python
print(results.params)
```

### 4.

```python
a0 = results.params['const']
a1 = results.params['RM']
result = a0 + a1*np.array([10.0])
print("{:.2f}".format(result[0]))
```

### 5.

```python
x_min_max = np.array([x['RM'].min(),x['RM'].max()])
y_min_max = a0 + a1 * x_min_max

plt.scatter(x["RM"], y)
plt.plot( x_min_max, y_min_max, c="red")

plt.xlabel("X['RM']")
plt.ylabel("MEDV")
plt.show()
```

### 6.

```python
variable = ["CRIM", "INDUS"]
y = df["MEDV"]
for s in variable:
    x  = df[[s]]
    x_c = sm.add_constant(x)
    model = sm.OLS(y, x_c)
    results = model.fit()

    a0 = results.params['const']
    a1 = results.params[s]

    x_min_max = np.array([x[s].min(), x[s].max()])    
    y_min_max = a0 + a1 * x_min_max

    plt.scatter(x[s], y)
    plt.plot( x_min_max, y_min_max, c="red")

    plt.xlabel(f"X['{s}']")
    plt.ylabel("MEDV")
    plt.show()
```
