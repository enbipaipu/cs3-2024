```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_csv("batting_npb2021.csv", sep=",")
print(df.shape)
print(df.info())
display(df.head())
```

### 説明変数と目的変数を分ける

```python
X = df[["Weight"]]
y = df["HomeRun"]

print("X:", X.shape)
display(X.head())
print("y:", y.shape)
print(y.head())
```

### 線形単回帰分析

```python
x_c = sm.add_constant(X)
display(X.head())
display(x_c.head())
```

```python
model = sm.OLS(y, x_c)
results = model.fit()
print(results.summary())
```

### 決定係数や自由度調整済み決定係数

```python
print("R2:", results.rsquared)
print("Adj R2:", results.rsquared_adj)
```

### 偏回路係数

```python
print(results.params)
```

### 回帰直線の描画

```python
a0=results.params["const"]
a1 = results.params["Weight"]
x_min = X["Weight"].min()
x_max = X["Weight"].max()
x_min_max = np.array([x_min, x_max])
y_min_max = a0+a1*x_min_max
plt.title('y = ({:.3f}) + ({:.3f}) * Weight'.format(a0, a1))
plt.scatter(X["Weight"], y)
plt.plot(x_min_max, y_min_max, c = "red")
plt.xlabel("Weight[kg]")
plt.ylabel("HomeRun")
plt.show()
```

### 得られたモデルを用いて、予測を行う

```python
x_test = np.array([70.0, 85.0, 100.0])
y_test = a0+a1*x_test
print(y_test)
```

```python
x_test_c = sm.add_constant(x_test, has_constant="add")
y_test = results.predict(x_test_c)
print(y_test)
```

```python
for c in ['Height', 'Steal', 'StrikeOut']:
    print(c, 'vs Weight')
    X = df[['Weight']]  # explanatory variable, 2D
    y = df[c]  # objective variable, 1D
    X_c = sm.add_constant(X)
    model = sm.OLS(y, X_c)
    results = model.fit()
    print('R2:', results.rsquared)
    print('Adj R2:', results.rsquared_adj)
    print(results.params)
    
    a0 = results.params['const']
    a1 = results.params['Weight']
    y_min_max = a0 + a1 * x_min_max

    plt.title('y = ({:.3f}) + ({:.3f}) * Weight'.format(a0, a1))
    plt.scatter(X['Weight'], y)
    plt.plot(x_min_max, y_min_max, c='red')
    plt.xlabel('Weight [kg]')
    plt.ylabel(c)
    plt.show()
```
