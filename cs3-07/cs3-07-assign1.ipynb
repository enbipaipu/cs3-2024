```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("boston_small.csv", sep=",", skiprows=8)
print(df.shape)
print(df.info())
display(df.head())
```

```python
df_corr = df.corr()
display("{:.2f}".format(df_corr.at["CRIM", "RM"]))
```

### 3.

```python
plt.figure(figsize=(10, 10))
sns.heatmap(df_corr, annot=True, fmt=".2f", cmap="bwr")
plt.show()
```

### 4.

```python
print(df.corrwith(df["MEDV"]))
```

### 5.

```python
plt.scatter(df['MEDV'], df["INDUS"])
plt.title("corr:{:.2f}".format(df_corr.at["MEDV", "INDUS"]))
plt.xlabel("MEDV")
plt.ylabel("INDUS")
plt.show()
```
