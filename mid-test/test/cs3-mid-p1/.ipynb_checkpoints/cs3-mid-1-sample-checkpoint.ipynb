```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_in = '2024-cs3-mid-1.csv'
df = pd.read_csv(csv_in,sep=",", skiprows=1, header=0)
print(df.shape)
display(df.head())
```

```python
print(df.shape)
print(df.info())
display(df.head())
```

```python
ser_c1  = df["c1"]
print(ser_c1.median())
```

```python
display(df.sort_values(by="c1", ascending=False).head())
```

```python
print(df["c6"].value_counts())
```

```python
df2 = df.drop(columns=["ID", "c4", "c5"])
display(df2.groupby('c6').mean())
```

```python
df["n_tot"] = df["c1"] + df["c2"] + df["c3"]
df3 = df.sort_values(by="n_tot", ascending=False)
display(df3.head())
```

```python
plt.hist(df3["n_tot"], bins=10)
plt.show()
```
