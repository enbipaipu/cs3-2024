```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter
```

```python
img = Image.open("cat_sample.jpg")
```

### 2

```python
print(img.format, img.size, img.width, img.height, img.mode)
```

### 3

```python
plt.imshow(img)
plt.show()
```

### 4

```python
data = np.array(img)
print(data.shape)
```

### 5

```python
print(data[200, 300])
```

### 6

```python
img_small = img.resize((30, 50))
plt.imshow(img_small)
plt.show()
```

### 7

```python
img_rot = img.rotate(60)
plt.imshow(img_rot)
plt.show()
```

### 8

```python
img_blur = img.filter(ImageFilter.GaussianBlur(radius=10))
plt.imshow(img_blur)
plt.show()
```

### 9

```python
img_crop=img.crop((100, 50, 300, 200))
plt.imshow(img_crop)
plt.show()
```

### 10

```python
img_gray = img.convert("L")
plt.imshow(img_gray, cmap="gray")
plt.show()
```

### 11

```python
img_out = "cat_gray.jpg"
img_gray.save(img_out, quality=95)
```
