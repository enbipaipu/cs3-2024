### random Forest Regressor (RFR)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from joblib import dump
```

```python
csv_in = "../cs3-08/student-por-gp.csv"

df = pd.read_csv(csv_in)
print(df.shape)
print(df.info())
display(df.head())
```

### 特徴量と正解の数値を分ける

```python
X = df.drop(columns='G3')  # features
y = df['G3']  # true values    
print('X:', X.shape)
display(X.head())
print('y:', y.shape)
print(y.head())
```

### カテゴリー変数の数値化

```python
print(X["studytime"].value_counts())
```

```python
X['studytime'] = X['studytime'].replace(
    {'<2 hours':'1', '2 to 5 hours':'2',
     '5 to 10 hours':'3', '>10 hours':'4',
    }
).astype('int')
display(X.head())
```

### ダミー変数化

```python
print(X["reason"].value_counts())
```

```python
X_dumm = pd.get_dummies(X, drop_first=True, dtype='uint8')
print('X_dumm:', X_dumm.shape)
display(X_dumm.head())
```

### 訓練データとテストデータに分割

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_dumm, y, test_size=0.3, random_state=13)
```

```python
print(X_train.shape)
print(y_train.shape)

print(X_test.shape)
print(y_test.shape)
```

### RFRの学習
- RandomForestRegressorでRFRの構造を定義する
- fit()で訓練用データと正解データを与えてRFRを学習

```python
rfr = RandomForestRegressor(n_estimators=500, max_depth=None, random_state=17)
```

```python
%%time
rfr.fit(X_train, y_train)
```

### 各特徴量の重要度

```python
ser_fi = pd.Series(rfr.feature_importances_, index=X_train.columns)
print(ser_fi.sort_values(ascending=False))
```

```python
plt.bar(X_train.columns, rfr.feature_importances_)
plt.ylabel("Importance")
plt.xticks(rotation=90)
plt.show()
```

### 予測を行う
- predict: 予測値の算出

```python
y_train_pred = rfr.predict(X_train)
```

### （発展）予測数値と実際の数値の対応

```python
y_min = np.append(y_train_pred, y_train).min()
y_max = np.append(y_train_pred, y_train).max()
y_margin = (y_max - y_min) * 0.1
y_min -= y_margin
y_max += y_margin

plt.scatter(y_train_pred, y_train, alpha=0.3)
plt.plot([y_min,y_max], [y_min,y_max])
plt.gca().set_aspect('equal', adjustable='box')
plt.xlim(y_min, y_max)
plt.ylim(y_min, y_max)
plt.title('Train data')
plt.xlabel('Pred')
plt.ylabel('True')
plt.show()
```

### モデルの当てはまりの良さの確認

```python
mse = mean_squared_error(y_train, y_train_pred)
print('MSE: ', mse)
print("RMSE for test data: ", np.sqrt(mse))
```

### 凡化性能の確認

```python
y_test_pred = rfr.predict(X_test)
```

```python
y_min = np.append(y_test_pred, y_test).min()
y_max = np.append(y_test_pred, y_test).max()
y_margin = (y_max - y_min) * 0.1
y_min -= y_margin
y_max += y_margin

plt.scatter(y_test_pred, y_test, alpha=0.3)
plt.plot([y_min,y_max], [y_min,y_max])
plt.gca().set_aspect('equal', adjustable='box')
plt.xlim(y_min, y_max)
plt.ylim(y_min, y_max)
plt.title('Test data')
plt.xlabel('Pred')
plt.ylabel('True')
plt.show()
```

```python
mse = mean_squared_error(y_test, y_test_pred)
print('MSE(平均2乗誤差): ', mse)
print("RMSE for test data(MSEの平方根): ", np.sqrt(mse))

tag = 'student-por'
model_file = 'rfr_{}.joblib'.format(tag)
dump(rfr, model_file)
```
